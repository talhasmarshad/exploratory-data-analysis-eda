```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # ✅ Needed for np.array()

# Data
data = {
    'Country': ['USA', 'China', 'India', 'Germany', 'Brazil', 'Japan', 'Russia', 'Canada', 'Australia', 'Mexico'],
    'GDP_per_Capita_USD': [59850, 10660, 1800, 46100, 10600, 38900, 10750, 41300, 51000, 9500],
    'Life_Expectancy_years': [78.54, 76.91, 69.42, 81.13, 75.88, 84.6, 73.34, 82.2, 82.5, 77.03],
    'Population_millions': [331, 1393, 1380, 83, 213, 126, 145, 38, 26, 128],
    'Color_Grading': [59850, 10660, 1800, 46100, 10600, 38900, 10750, 41300, 51000, 9500]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Extract columns
gdp_cap = df['GDP_per_Capita_USD']
life_exp = df['Life_Expectancy_years']
pop = df['Population_millions']
color = df['Color_Grading']

# Convert population to NumPy and scale
np_pop = np.array(pop) * 2  # multiply for better visibility

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(gdp_cap, life_exp, s=np_pop, c=color, cmap='viridis', alpha=0.7, edgecolors='black')

# Set log scale for x-axis
plt.xscale('log')

# Labels, title, and ticks
plt.xlabel('GDP per Capita [USD]')
plt.ylabel('Life Expectancy [years]')
plt.title('🌍 Country Life Expectancy vs GDP per Capita')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Optional: color bar
plt.colorbar(label='GDP per Capita [USD]')

# Show plot
plt.show()
```
